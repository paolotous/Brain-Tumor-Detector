```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('test_images/0 (239).jpg', 0)
height, width = img.shape
plt.imshow(img, cmap='gray')
```

```python
imgFlattened = img.reshape(height*width, 1)
print(imgFlattened.shape)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6)
model.fit(imgFlattened)
labels = model.labels_
labels2D = labels.reshape(height, width)

print(img[200:210, 200:210])
print("-------------------------------")
print(labels2D[200:210, 200:210])
```

```python
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
import joblib

joblib.dump(model, 'brain_cluster.sav')
```
